### Data Importing and Cleaning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../datasets/train.csv', keep_default_na=False, na_values=[''])
pd.options.display.max_columns = 99
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
#Cols with nulls 
df.isnull().sum().sort_values(ascending=False).head(22)
```

```python
#dropping nan's
df = df.dropna()
```

```python
df.head()
```

```python
#Checking data types of cols; all seem to be the right type!
df.info()
```

```python
#Removing first 2 columns; unnecessary:
df = df.drop(columns=['Id','PID'])
```

```python
#Exporting Cleaned DataFrame
df.to_csv('../datasets/cleaned_housing_data.csv')
```
